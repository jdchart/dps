## 4. Feature Extraction Selfmade
Now we can use the [dps python package](https://github.com/jdchart/dps.git) to extract the various features. We shall do this for each file.

Before we process everything, let's just take a closer look at one instance in order to understand what's happening under the hood.

```python
# Charger toutes les courbes DPS à partir des fichiers JSON
all_dps_curves = {}  # Dictionnaire pour stocker les courbes avec leurs noms

setup_window = 480  # Paramètre `setup_window`
hop_size = 10        # Paramètre `hop_size`

print(f"Processing {len(analysis_files)} JSON files...")

for file in analysis_files:
    # Charger le fichier d'analyse
    speech_recognition = dps.SpeechAnalysis(file, fps=32)
    
    # Calculer la courbe DPS
    dps_curve = speech_recognition.get_dps_feature_curve(setup_window, hop_size)
    
    # Extraire le nom du fichier pour l'utiliser comme clé
    file_name = os.path.basename(file)
    
    # Stocker la courbe
    all_dps_curves[file_name] = dps_curve

print(f"Processed {len(all_dps_curves)} files.")
```

```python
plt.figure(figsize=(12, 8))

fps = 32  # Frame rate utilisé dans `speech_recognition`

# Pour chaque fichier et sa courbe DPS
for file_name, dps_curve in all_dps_curves.items():
    # Créer les numéros de frame et convertir en temps (MM:SS)
    frame_numbers = np.arange(len(dps_curve))
    times_in_seconds = frame_numbers / fps  # Conversion des frames en secondes

    # Tracer la courbe avec une légende
    plt.plot(times_in_seconds, dps_curve, label=file_name, drawstyle='steps-post')

# Ajouter les labels et le titre
plt.xlabel('Time (MM:SS)')
plt.ylabel('DPS')
plt.title('DPS Curves for All JSON Files')

# Formater les étiquettes en MM.SS
def format_time(seconds):
    minutes = int(seconds // 60)
    seconds = int(seconds % 60)
    return f"{minutes:02}:{seconds:02}"

# Ajuster les étiquettes de l'axe des x
plt.xticks(
    ticks=np.arange(0, max(times_in_seconds), 300),  # Un tick toutes les 5 minutes
    labels=[format_time(t) for t in np.arange(0, max(times_in_seconds), 300)],
    rotation=45,  # Inclinaison des étiquettes
    ha='right'    # Alignement
)

# Ajouter une légende
plt.legend(loc='upper right', bbox_to_anchor=(1.25, 1.0), fontsize='small')

# Afficher le graphique
plt.tight_layout()
plt.show()
```

## 7. Moyenne et ecart type


```python
# Méthode C : Moyenne et écart-type de toutes les courbes

# Trouver la longueur minimale pour aligner les courbes
min_len = min(len(c) for c in all_dps_curves.values())

# Tronquer les courbes à cette longueur
truncated_curves = np.array([curve[:min_len] for curve in all_dps_curves.values()])
mean_curve = np.mean(truncated_curves, axis=0)
std_curve = np.std(truncated_curves, axis=0)
times = np.arange(min_len) / fps

# Affichage
plt.figure(figsize=(12, 6))
plt.plot(times, mean_curve, label='Mean WPS', color='blue')
plt.fill_between(times, mean_curve - std_curve, mean_curve + std_curve, color='blue', alpha=0.3, label='±1 std')
plt.title("Average WPS Curve with Standard Deviation")
plt.xlabel("Time (s)")
plt.ylabel("WPS")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np

# Bins pour la répartition des valeurs de WPS
bins = [0, 1, 2, 3, 5]
bin_labels = ["[0–1[", "[1–2[", "[2–3[", "[3–5]"]

stats_summary = []

for file_name, curve in all_dps_curves.items():
    data = np.array(curve, dtype=np.float64)
    
    # Remplacer les NaN par 0
    data = np.nan_to_num(data, nan=0.0)

    duration = len(data) / fps

    # Calculs statistiques
    stats = {
        "file": file_name,
        "duration (s)": round(duration, 2),
        "mean": np.mean(data),
        "std": np.std(data),
        "min": np.min(data),
        "max": np.max(data),
        "median": np.median(data),
        "p25": np.percentile(data, 25),
        "p75": np.percentile(data, 75)
    }

    # Répartition par classes
    bin_counts, _ = np.histogram(data, bins=bins)
    for label, count in zip(bin_labels, bin_counts):
        stats[f"count {label}"] = count
        stats[f"perc {label}"] = count / len(data) * 100

    stats_summary.append(stats)

# Création du DataFrame
stats_df = pd.DataFrame(stats_summary)
stats_df = stats_df.round(2)
stats_df.sort_values("mean", inplace=True)

# Affichage
display(stats_df)

# Export si souhaité
stats_df.to_csv("wps_statistics_summary_clean.csv", index=False)
print("Statistiques exportées dans 'wps_statistics_summary_clean.csv'")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Colonnes de proportions à visualiser
perc_cols = [col for col in stats_df.columns if col.startswith("perc ")]

# Réindexer le DataFrame dans l’ordre initial des fichiers
file_order = list(all_dps_curves.keys())
heatmap_data = stats_df.set_index("file").loc[file_order, perc_cols]

# Affichage
plt.figure(figsize=(12, 8))
sns.heatmap(heatmap_data, annot=True, fmt=".1f", cmap="Blues", cbar_kws={'label': '% of WPS samples'})
plt.title("WPS Distribution Across Bins by File (%)")
plt.ylabel("File")
plt.xlabel("WPS Ranges")
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Données numériques uniquement (exclure nom du fichier)
feature_cols = ['mean', 'std', 'min', 'max', 'median', 'p25', 'p75'] + perc_cols
X = stats_df[feature_cols].fillna(0)  # sécurité supplémentaire

# Standardisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Clustering
kmeans = KMeans(n_clusters=3, random_state=0)
labels = kmeans.fit_predict(X_scaled)

# Réduction en 2D avec PCA pour visualisation
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='Set2', s=100)
plt.title("Clustering of Performances Based on WPS Statistics")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.grid(True)

# Légendes des points
for i, name in enumerate(stats_df["file"]):
    plt.annotate(name[:12], (X_pca[i, 0], X_pca[i, 1]), fontsize=8, alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt

# Réduction avec t-SNE
tsne = TSNE(n_components=2, perplexity=5, random_state=0)
X_tsne = tsne.fit_transform(X_scaled)

plt.figure(figsize=(10, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='Set2', s=100)
plt.title("t-SNE projection of WPS Statistics")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.grid(True)

for i, name in enumerate(stats_df["file"]):
    plt.annotate(name[:12], (X_tsne[i, 0], X_tsne[i, 1]), fontsize=8, alpha=0.7)

plt.tight_layout()
plt.show()
```

## 7.2. SANS les 0


```python
stats_nonzero = []

plt.figure(figsize=(12, 8))

for file_name, curve in all_dps_curves.items():
    data = np.array(curve, dtype=np.float64)
    data = np.nan_to_num(data, nan=0.0)

    # Exclure tous les points où WPS = 0
    non_zero_data = data[data > 0]

    if len(non_zero_data) == 0:
        print(f"Skipping {file_name} (only zeros)")
        continue

    duration = len(non_zero_data) / fps

    stats = {
        "file": file_name,
        "duration (non-zero) (s)": round(duration, 2),
        "mean": np.mean(non_zero_data),
        "std": np.std(non_zero_data),
        "min": np.min(non_zero_data),
        "max": np.max(non_zero_data),
        "median": np.median(non_zero_data),
        "p25": np.percentile(non_zero_data, 25),
        "p75": np.percentile(non_zero_data, 75)
    }

    bin_counts, _ = np.histogram(non_zero_data, bins=bins)
    for label, count in zip(bin_labels, bin_counts):
        stats[f"count {label}"] = count
        stats[f"perc {label}"] = count / len(non_zero_data) * 100

    stats_nonzero.append(stats)

    # Tracer la courbe nettoyée
    time = np.arange(len(data)) / fps
    plt.plot(time, data, label=file_name, drawstyle='steps-post', alpha=0.5)

# Affichage de toutes les courbes
plt.title("DPS Curves (WPS) with Zero Values Removed")
plt.xlabel("Time (s)")
plt.ylabel("WPS")
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0), fontsize='small')
plt.tight_layout()
plt.show()

# Résumé statistique
stats_df_nz = pd.DataFrame(stats_nonzero).round(2)
display(stats_df_nz)
```

## 8. normalize + windows


```python
import numpy as np
import pandas as pd

N_WINDOWS = 150
fps = 32
bins = [0, 1, 2, 3, 5]
bin_labels = ["[0–1[", "[1–2[", "[2–3[", "[3–5]"]

window_stats = []

for file_name, curve in all_dps_curves.items():
    data = np.array(curve, dtype=np.float64)
    data = np.nan_to_num(data, nan=0.0)

    total_len = len(data)
    points_per_window = total_len // N_WINDOWS

    for w in range(N_WINDOWS):
        start = w * points_per_window
        end = (w + 1) * points_per_window if w < N_WINDOWS - 1 else total_len
        segment = data[start:end]
        segment_nonzero = segment[segment > 0]

        # Temps de début et de fin (en secondes)
        start_time = start / fps
        end_time = end / fps

        # Statistiques de base
        stats = {
            "file": file_name,
            "window": w,
            "start_time_s": round(start_time, 2),
            "end_time_s": round(end_time, 2),
            "start_time_mn": f"{int(start_time // 60):02}:{int(start_time % 60):02}",
            "end_time_mn": f"{int(end_time // 60):02}:{int(end_time % 60):02}",
            "count_words": len(segment_nonzero),
            "mean": np.mean(segment_nonzero) if len(segment_nonzero) > 0 else 0,
            "std": np.std(segment_nonzero) if len(segment_nonzero) > 0 else 0,
            "min": np.min(segment_nonzero) if len(segment_nonzero) > 0 else 0,
            "max": np.max(segment_nonzero) if len(segment_nonzero) > 0 else 0
        }

        # Répartition par bins
        bin_counts, _ = np.histogram(segment_nonzero, bins=bins)
        for label, count in zip(bin_labels, bin_counts):
            stats[f"count {label}"] = count
            stats[f"perc {label}"] = (count / len(segment_nonzero) * 100) if len(segment_nonzero) > 0 else 0

        window_stats.append(stats)

# Création du DataFrame final
df_windows = pd.DataFrame(window_stats).round(2)
display(df_windows.head(20))
```

```python
import numpy as np
import pandas as pd

# Recharger le CSV du chronométrage manuel
stopwatch_csv_path = r"C:\Users\theoh\Documents\Fac\Thèse\Donnees\DPS_LMI\Fusion\operate1curve\LMI_Stopwatch_20230509.csv"
df_stopwatch = pd.read_csv(stopwatch_csv_path, decimal=',')


# Définir la durée cible du filage (en secondes)
total_duration = df_stopwatch['end_time_s'].max()

# Créer un signal "continu" à la seconde près (rempli de NaN par défaut)
wps_signal = np.full(int(np.ceil(total_duration)), np.nan)

# Remplir les valeurs du signal selon chaque séquence
for _, row in df_stopwatch.iterrows():
    start = int(np.floor(row['start_time_s']))
    end = int(np.ceil(row['end_time_s']))
    wps_signal[start:end] = row['wps_mean']

# Nombre de fenêtres (doit matcher df_windows pour heatmap)
N_WINDOWS = 150
samples_per_window = len(wps_signal) // N_WINDOWS

manual_window_stats = []

for i in range(N_WINDOWS):
    start = i * samples_per_window
    end = (i + 1) * samples_per_window if i < N_WINDOWS - 1 else len(wps_signal)
    segment = wps_signal[start:end]
    segment_non_nan = segment[~np.isnan(segment)]
    
    mean_wps = np.mean(segment_non_nan) if len(segment_non_nan) > 0 else np.nan

    manual_window_stats.append({
        'file': 'Manual_Stopwatch',
        'window': i,
        'mean': round(mean_wps, 2)
    })

# Convertir en DataFrame
df_manual_windows = pd.DataFrame(manual_window_stats)
```

```python
# Exemple : événements à relier
scene_events = pd.DataFrame([
    {"file": "LMI_20230512...", "time_s": 14040, "label": "entrée micro"},
    {"file": "LMI_20230512...", "time_s": 1980, "label": "scène collective"},
    {"file": "LMI_20230510...", "time_s": 1200, "label": "reverb intense"},
])
```

```python
# Associer chaque événement à une fenêtre
event_windows = []

for _, event in scene_events.iterrows():
    matches = df_windows[
        (df_windows["file"].str.startswith(event["file"][:20])) &
        (df_windows["start_time_s"] <= event["time_s"]) &
        (df_windows["end_time_s"] > event["time_s"])
    ]
    if not matches.empty:
        row = matches.iloc[0].copy()
        row["event_label"] = event["label"]
        event_windows.append(row)

# Créer le DataFrame annoté
df_event_windows = pd.DataFrame(event_windows)
```

```python
import matplotlib.pyplot as plt

# Ordre des fichiers d'origine
ordered_files = list(all_dps_curves.keys())

plt.figure(figsize=(14, len(ordered_files) * 0.5))

for i, file in enumerate(ordered_files):
    file_windows = df_windows[df_windows["file"] == file]
    for _, row in file_windows.iterrows():
        color = cluster_colors_hex[int(row["cluster"])]
        plt.plot([row["start_time_s"] / 60, row["end_time_s"] / 60],
                 [i, i], color=color, linewidth=6)

# Annotations éventuelles
if 'df_event_windows' in locals():
    for _, row in df_event_windows.iterrows():
        try:
            y_index = ordered_files.index(row["file"])
            x_pos = (row["start_time_s"] + row["end_time_s"]) / 2 / 60
            plt.text(x_pos, y_index + 0.3, row["event_label"],
                     fontsize=8, ha="center", va="bottom", color="black")
        except ValueError:
            continue

plt.yticks(range(len(ordered_files)), [f[:12] for f in ordered_files])
plt.xlabel("Time (minutes)")
plt.title("Timeline des clusters rythmiques (par fenêtre)")
plt.grid(True, axis='x', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
